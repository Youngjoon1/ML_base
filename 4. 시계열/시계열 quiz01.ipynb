```python
rental_num = df[df['대여거치대'].isnull()]['대여소번호']
rental_num
```

```python
rental_num = rental_num.unique()
rental_num
```

```python
df[df['대여소번호']==646]['대여거치대']
```

```python
print(df[df['대여소번호']==646]['대여거치대'].mean())
round(df[df['대여소번호']==646]['대여거치대'].mean(),0)
```

```python
rental_dic = {}
for number in rental_num:
    rental_dic[number] = round(df[df['대여소번호']==number]['대여거치대'].mean(),0);
rental_dic
```

```python
df.loc[df['대여소번호']==646,'대여거치대']=df[df['대여소번호']==646]['대여거치대'].fillna(round(df[df['대여소번호']==646]['대여거치대'].mean(),0))
```
