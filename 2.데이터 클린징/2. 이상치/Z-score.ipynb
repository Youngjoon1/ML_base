```python
import pandas as pd
df = pd.read_csv('data/food.csv')
df
```

```python
mean = df['food_a'].mean()
mean
```

```python
std = df['food_a'].std()
std
```

```python
threshold = 3
outlier = []
for j in df.columns:   
    for i in df[j]:
        z = (i-mean)/std
        if z > threshold:
            outlier.append(i)
   
```
